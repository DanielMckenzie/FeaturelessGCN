```python
from src.get_data import load_benchmark, load_synthetic
from src.normalization import get_adj_feats, normalization, sparse_mx_to_torch_sparse_tensor
from src.args import get_args 
from src.models import get_model
import torch.optim as optim
import torch
import torch.nn.functional as F
import time
from src.utils import accuracy
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
from src.plots import plot_feature, plot_heatmap, plot_graph, plot_degree


from src.utils import LDA_loss
```

```python
# load dataset
# all tensor, dense
dataset_name = 'SmallGap'#same_feats_1D_sparse, same_feats_2D_dense
split_index = 1
adj, feats, labels, idx_train, idx_val, idx_test = load_synthetic(dataset_name, split_index)
# adj, feats, labels, idx_train, idx_val, idx_test = load_benchmark(dataset_name)
```

```python
nb_class = (torch.max(labels) + 1).numpy()
Y_onehot =  torch.zeros(labels.shape[0], nb_class).scatter_(1, labels.unsqueeze(-1), 1)

nb_each_class_train = torch.sum(Y_onehot[idx_train], dim = 0)
nb_each_class_inv_train = torch.tensor(np.power(nb_each_class_train.numpy(), -1).flatten())
nb_each_class_inv_mat_train = torch.diag(nb_each_class_inv_train)

nb_each_class_val = torch.sum(Y_onehot[idx_val], dim = 0)
nb_each_class_inv_val = torch.tensor(np.power(nb_each_class_val.numpy(), -1).flatten())
nb_each_class_inv_mat_val = torch.diag(nb_each_class_inv_val)

nb_each_class_test = torch.sum(Y_onehot[idx_test], dim = 0)
nb_each_class_inv_test = torch.tensor(np.power(nb_each_class_test.numpy(), -1).flatten())
nb_each_class_inv_mat_test = torch.diag(nb_each_class_inv_test)

nb_each_class_all = torch.sum(Y_onehot, dim = 0)
nb_each_class_inv_all = torch.tensor(np.power(nb_each_class_all.numpy(), -1).flatten())
nb_each_class_inv_mat_all = torch.diag(nb_each_class_inv_all)
```

```python
GCN = 'GCN'
args = get_args(model_opt = GCN, dataset = dataset_name)
adj_GCN, feats_GCN = get_adj_feats(adj = adj, feats = feats, model_opt = GCN, degree = args.degree)


SGC = 'SGC'
args = get_args(model_opt = SGC, dataset = dataset_name)
adj_SGC, feats_SGC = get_adj_feats(adj = adj, feats = feats, model_opt = SGC, degree = args.degree)

GFNN = 'GFNN'
args = get_args(model_opt = GFNN, dataset = dataset_name)
adj_GFNN, feats_GFNN = get_adj_feats(adj = adj, feats = feats, model_opt = GFNN, degree = args.degree)

GFN = 'GFN'
args = get_args(model_opt = GFN, dataset = dataset_name)
adj_GFN, feats_GFN = get_adj_feats(adj = adj, feats = feats, model_opt = GFN, degree = args.degree)

GIN = 'GIN'
args = get_args(model_opt = GIN, dataset = dataset_name)
adj_GIN, feats_GIN = get_adj_feats(adj = adj, feats = feats, model_opt = GIN, degree = args.degree)
```

```python
AGNN = 'AGNN'
args = get_args(model_opt = AGNN, dataset = dataset_name)
adj_AGNN, feats_AGNN = get_adj_feats(adj = adj, feats = feats, model_opt = AGNN, degree = args.degree)
weight = [0.1678, 0.0056, 0.0156, 0.0218, 0.0073, 0.0201, 0.7084, 0.0069, 0.0185, 0.0280]#smallGap
# weight = [0.1718, 0.0076, 0.0196, 0.0278, 0.0083, 0.0221, 0.6844, 0.0079, 0.0205, 0.0300]#smallGap
# weight = [4.2743e-04, 1.8011e-02, 4.6940e-01, 5.0647e-01, 8.4062e-04, 4.7787e-04,
#         5.2185e-04, 2.6306e-03, 7.2868e-04, 4.9805e-04]#samefeats
adj_AGNN_sum=weight[0] * torch.eye(adj_AGNN[0].shape[1])
for i in range(len(adj_AGNN)):
    adj_AGNN_sum = adj_AGNN_sum + weight[i+1] * adj_AGNN[i]
    

```

```python
# plot

print("-------------GCN-------------")
feats_GCN = torch.spmm(adj_GCN, feats_GCN)
LDA_loss_GCN = LDA_loss(feats_GCN, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_GCN)
plot_feature(feats_GCN.numpy(), labels.numpy(), Y_onehot.shape[1]) 



print("-------------GIN-------------")
feats_GIN = torch.spmm(adj_GIN, feats_GIN)
LDA_loss_GIN = LDA_loss(feats_GIN, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_GIN)
plot_feature(feats_GIN.numpy(), labels.numpy(), Y_onehot.shape[1]) 


print("-------------SGC-------------")
feats_SGC = torch.spmm(adj_SGC, feats_SGC)
LDA_loss_SGC = LDA_loss(feats_SGC, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_SGC)
plot_feature(feats_SGC.numpy(), labels.numpy(), Y_onehot.shape[1])    
    
    
print("-------------GFN-------------")
feats_GFN = torch.spmm(adj_GFN, feats_GFN)
LDA_loss_GFN = LDA_loss(feats_GFN, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_GFN)
plot_feature(feats_GFN.numpy(), labels.numpy(), Y_onehot.shape[1]) 


print("-------------GFNN-------------")
feats_GFNN = torch.spmm(adj_GFNN, feats_GFNN)
LDA_loss_GFNN = LDA_loss(feats_GFNN, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_GFNN)
plot_feature(feats_GFNN.numpy(), labels.numpy(), Y_onehot.shape[1]) 


print("-------------AGNN-------------")
feats_AGNN = torch.spmm(adj_AGNN_sum, feats_AGNN)
LDA_loss_AGNN = LDA_loss(feats_AGNN, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print('score is: ',LDA_loss_AGNN)
plot_feature(feats_AGNN.numpy(), labels.numpy(), Y_onehot.shape[1]) 
    
```
