```python
# load dataset
# all tensor, dense
dataset_name = 'SameFeats'#same_feats_1D_sparse, same_feats_2D_dense
split_index = 1
adj, feats, labels, idx_train, idx_val, idx_test = load_synthetic(dataset_name, split_index)
# adj, feats, labels, idx_train, idx_val, idx_test = load_benchmark(dataset_name)
```

```python
nb_class = (torch.max(labels) + 1).numpy()
Y_onehot =  torch.zeros(labels.shape[0], nb_class).scatter_(1, labels.unsqueeze(-1), 1)

nb_each_class_train = torch.sum(Y_onehot[idx_train], dim = 0)
nb_each_class_inv_train = torch.tensor(np.power(nb_each_class_train.numpy(), -1).flatten())
nb_each_class_inv_mat_train = torch.diag(nb_each_class_inv_train)

nb_each_class_val = torch.sum(Y_onehot[idx_val], dim = 0)
nb_each_class_inv_val = torch.tensor(np.power(nb_each_class_val.numpy(), -1).flatten())
nb_each_class_inv_mat_val = torch.diag(nb_each_class_inv_val)

nb_each_class_test = torch.sum(Y_onehot[idx_test], dim = 0)
nb_each_class_inv_test = torch.tensor(np.power(nb_each_class_test.numpy(), -1).flatten())
nb_each_class_inv_mat_test = torch.diag(nb_each_class_inv_test)

nb_each_class_all = torch.sum(Y_onehot, dim = 0)
nb_each_class_inv_all = torch.tensor(np.power(nb_each_class_all.numpy(), -1).flatten())
nb_each_class_inv_mat_all = torch.diag(nb_each_class_inv_all)
```

```python
# print score for all

raw_feats = feats
A = adj.numpy()
LDA_loss_train = LDA_loss(feats, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
print("method: identity", "propogate times: 0", "score is: ", LDA_loss_train)
plot_feature(feats.numpy(),labels.numpy(),Y_onehot.shape[1])

for method in ['row','col','sym']:
    normed_adj = normalization(A = A, norm_method = method, num_power = 1)
    normed_adj = sparse_mx_to_torch_sparse_tensor(normed_adj)
    feats = raw_feats
    for i in range(3):
        feats = torch.spmm(normed_adj, feats) 
        LDA_loss_train = LDA_loss(feats, Y_onehot, nb_each_class_inv_mat_all, norm_or_not = True)
        print("method: ",method, "propogate times: ", i+1, "score is: ", LDA_loss_train)
        plot_feature(feats.numpy(),labels.numpy(),Y_onehot.shape[1])
    
        
        
```
